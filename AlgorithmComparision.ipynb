```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('CropRecommenderUpdated.csv')
```

Data Preprocessing 

```python
from sklearn import preprocessing 
label_encoder=preprocessing.LabelEncoder()
df['label']=label_encoder.fit_transform(df['label'])
df['label'].unique()
```

```python
X=df.drop('label', axis=1)
Y=df[['label']]
```

```python
import sklearn
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
# splitting the data for training and testing
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2,random_state=42)
```

Standardization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

Decision Tree

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier(criterion="gini", random_state=42,max_depth=3, min_samples_leaf=5)     #model implemented for Gini criterion.
dt_model.fit(x_train,y_train)
```

```python
y_predicted = dt_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import classification_report
```

```python
print('Validation Accuracy:', accuracy_score(y_test, y_predicted))  # performance measure for Gini.
```

```python
print(classification_report(y_test,y_predicted))
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
confusion_matrix = metrics.confusion_matrix(y_test,y_predicted)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix)
cm_display.plot()
plt.show()
```

```python
clf = tree.DecisionTreeClassifier(random_state=0)
clf = clf.fit(X,Y)
```

```python
dt_model = DecisionTreeClassifier(criterion="entropy", random_state=42,max_depth=3, min_samples_leaf=5)      #model implemented for Entropy criterion.
dt_model.fit(x_train,y_train)
```

```python
y_predicted = dt_model.predict(x_test)
```

```python
print('Validation Accuracy:', accuracy_score(y_test, y_predicted))     # performance measure for Entropy.
```

```python
print(classification_report(y_test,y_predicted))
```

```python
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix)
cm_display.plot()
plt.show()
```

Multivariate Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
# Create the linear regression model
reg = LinearRegression()

# Train the model
reg.fit(x_train, y_train)

# Predict the target values for the test data
y_predicted = reg.predict(x_test)

# Evaluate the model
mse = np.mean((y_test - y_predicted)**2)
r2 = reg.score(x_test, y_test)

print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
```
