```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Crop_recommendation.csv')
```

```python
Q1 = df['N'].quantile(0.25)
Q3 = df['N'].quantile(0.75)
```

```python
print("75th quartile: ",Q3)
print("25th quartile: ",Q1)
```

```python
IQR = Q3 - Q1
print ("IQR: ",IQR)
```

```python
upper_limit=Q3 + 1.5*IQR
lower_limit=Q1 - 1.5*IQR
```

```python
df[df['N'] > upper_limit].count()
```

```python
Q1 = df['P'].quantile(0.25)
Q3 = df['P'].quantile(0.75)

IQR = Q3 - Q1

upper_limit=Q3 + 1.5*IQR
lower_limit=Q1 - 1.5*IQR

df[df['P'] > upper_limit].count()
df[df['P'] < lower_limit].count()
```

```python
Q1 = df['temperature'].quantile(0.25)
Q3 = df['temperature'].quantile(0.75)

IQR = Q3 - Q1

upper_limit=Q3 + 1.5*IQR
lower_limit=Q1 - 1.5*IQR

new_df1 = df[(df['temperature'] < upper_limit) & (df['temperature']>lower_limit)]
new_df1.shape
```

```python
Q1 = df['humidity'].quantile(0.25)
Q3 = df['humidity'].quantile(0.75)

IQR = Q3 - Q1

upper_limit=Q3 + 1.5*IQR
lower_limit=Q1 - 1.5*IQR

new_df2 = df[(df['humidity'] < upper_limit) & (df['humidity']>lower_limit)]
new_df2.shape
```

```python
Q1 = df['ph'].quantile(0.25)
Q3 = df['ph'].quantile(0.75)

IQR = Q3 - Q1

upper_limit=Q3 + 1.5*IQR
lower_limit=Q1 - 1.5*IQR

new_df3 = df[(df['ph'] < upper_limit) & (df['ph']>lower_limit)]
new_df3.shape
```

```python
print(new_df1)
```

```python
dfx=pd.merge(new_df1,new_df2,on='Id')
```

```python
df_upd=pd.merge(dfx,new_df3,on='Id')
```
