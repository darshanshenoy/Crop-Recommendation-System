Nitrogen Content

```python
N=df.iloc[:,0]   #Phosphorous column in dataframes.
Nmin=min(N)  #0
Nmax=max(N)  #140
print(Nmin,Nmax)
```

```python
plt.boxplot(N)  
plt.show()
```

```python
sorted(N)
Q1=N.quantile(0.25)
Q3=N.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(N))
print(z)
#it is difficult to say which data point is an outlier
```

```python
#defining threshold
threshold = 3
print(np.where(z > 3)) 
```
